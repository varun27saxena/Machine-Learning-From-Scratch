```python
# Linear Regression OLS implementation
class LinearRegressionOLS:
    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, X_train, y_train):

        X_train = np.asarray(X_train).ravel()
        y_train = np.asarray(y_train).ravel()

        self.m = np.sum((X_train - X_train.mean()) *
                        (y_train - y_train.mean())) / \
                 np.sum((X_train - X_train.mean()) ** 2)

        self.b = y_train.mean() - self.m * X_train.mean()

        print("Coef:", self.m)
        print("Intercept:", self.b)

    def predict(self, X_test):
        X_test = np.asarray(X_test).ravel()
        return self.m * X_test + self.b
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns = ['Salary']),df['Salary'],test_size = 0.2,random_state = 42)
```
